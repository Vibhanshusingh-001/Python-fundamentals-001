# gc content from fasta file 

```python
import os
from Bio import SeqIO
import csv

def calculate_gc_content(fasta_file):
    """
    Calculate GC content of sequences in a FASTA file.
    """
    total_length = 0
    gc_count = 0
    
    for record in SeqIO.parse(fasta_file, "fasta"):
        sequence = record.seq.upper()
        total_length += len(sequence)
        gc_count += sequence.count("G") + sequence.count("C")
    
    if total_length == 0:  # Handle empty sequences
        return 0.0
    
    return round((gc_count / total_length) * 100, 2)

def process_fasta_files(input_dir, output_file):
    """
    Process all FASTA files in a directory to calculate GC content.
    """
    results = []
    
    # Iterate over all files in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".fasta"):  # Check for FASTA files
            filepath = os.path.join(input_dir, filename)
            gc_content = calculate_gc_content(filepath)
            results.append((filename, gc_content))
    
    # Save results to a CSV file
    with open(output_file, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Filename", "GC_Content (%)"])  # Header row
        writer.writerows(results)
    
    print(f"GC content results saved to {output_file}")

# Input directory containing FASTA files
input_directory = "C:/Users/vibha/Desktop/biopython/seq"

# Output CSV file to save results
output_csv = "gc_content_results.csv"

# Run the processing function
process_fasta_files(input_directory, output_csv)
```

```python
from Bio import SeqIO
file_path="C:/Users/vibha/Desktop/biopython/seq/sequence.gb"
gb_obj=SeqIO.read(file_path,"gb")
genes=[]
```

```python
for feature in gb_obj.features:
    if feature.type=="gene":
        genes.append(feature)
```

```python
len(genes)
gene_of_interest="rpoB"
hits=[]
for gene in genes:
    if "gene" in gene.qualifiers.keys():
        if gene_of_interest == gene.qualifiers["gene"][0]:
            hits.append(gene)
            print("gene found")
```
